```python
import numpy as np 
r1=np.random.randint(0,255,(500,500,3))
plt.imshow(r1)
```

```python
r2=np.random.randint(0,255,(50,50))
plt.imshow(r2,cmap='gray')
```

```python
img=np.ones((50,50,3),dtype='uint8')*255
img[:25,:25,1:]= 0 #    img[:25,:25,1:]*0
img[25:,:25,:2]=0 #     img[25:,:25,:2]*0
img[:25,25:,::2]= 0 #    img[:25,25:,::2]*0
img[25:,25:,2:]= 0 #       img[25:,25:,2:]*0



plt.imshow(img)
```

```python
plt.imsave('squarebox.png',img)
```

```python
skimage.io.imsave('squarebox.png',img)
```

```python
cv2.imwrite('squarebox.png',img[:,:,::-1])
```
